# Limpieza de valores redundantes

#### Referencias al municipio y al departamente son redundantes, ya que todos los datos pertenecen al mismo departamento y al mismo municipio, es decir, departamento de Antioquia, municipio de Medellin en Colombia. 

```python
dfCol_2.ciudad_municipio_nom.value_counts()
```

```python
dfCol_2.departamento_nom.value_counts()
```

```python
col_Ant=['departamento_nom', 'ciudad_municipio_nom']
dfCol_3=dfCol_2.drop(labels=col_Ant, axis=1)
dfCol_3.columns
```

# Recodificar variables categoricas

```python
alldata=[] # lista para guardar resultados de One hot Encoding
alldata1=[] # lista para guardar resultados de Label Encoder 
variables=[] # guardado de las variables
for i in dfCol_3.columns:
  if dfCol_3[i].dtype == 'O': # Nos quedamos con las que sean tipo object.
    if len(dfCol_3[i].unique()) ==2: # Si solo tienen dos categorias aplicar One Hot Encoding
      print('Variable 2 categorias:',i)
      y= pd.get_dummies(dfCol_3[i])
      alldata.append(y) # ir agregando las columnas resultantes en alldata
      variables.append(i)
    else:
      pass
    print('--------------------')
    if len(dfCol_3[i].unique()) >2: # Si las categorias tienen más de 2 categorias se aplica LabelEncoder
      print('Variable >2 categorias:',i)
      z= pd.DataFrame()
      z[i]=dfCol_3[i].astype('category').cat.codes # Aplicar Label Encoder
      alldata1.append(z) # Agregar a la lista alldata1
      variables.append(i)
  if dfCol_3[i].dtype == 'float64': # Si la columna es numerica entonces normalizar z score
    dfCol_3[i]=(dfCol_3[i]-dfCol_3[i].mean())/(dfCol_3[i].std())
data_y=pd.concat(alldata, axis=1) # concatenar las columnas obtenidas One hot Encoding
data_z=pd.concat(alldata1, axis=1) # concatenar las columnas de Label Encoder
dfCol_Ok= pd.concat([data_y,data_z,dfCol_3],axis=1) # Unir todos los datasets One Hot Encoding, LabelEncoder, Z score
dfCol_Ok=dfCol_Ok.drop(labels=variables, axis=1) # Borrar las columnas ya recodificadas
dfCol_Ok.head() # mostrar que quedo 
```

#### La columna departamento_nom no presenta ningún tipo de variación relevante. Todo los datos son del departamento de Antioquia en Colombia, por tanto, la referencia es redundantne. Se decide eliminarla para no tener datos tipo object dentro de la base.
